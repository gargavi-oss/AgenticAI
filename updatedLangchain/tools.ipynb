```python
import os
from dotenv import load_dotenv
load_dotenv()

os.environ["GOOGLE_API_KEY"]=os.getenv("GOOGLE_API_KEY")
from langchain.chat_models import init_chat_model
```

```python
model = init_chat_model("google_genai:gemini-3.1-flash-lite")
model
```

```python
from langchain.tools import tool

@tool
def getWeather(loc:str)->str:
    """Get weather at the location"""
    return f"It is sunny at {loc}"


model_with_tools= model.bind_tools([getWeather])
```

```python
response = model_with_tools.invoke("How is the weather in india?")
```

```python
messages = [{"role":"user","content":"What is the weather in delhi?"}]
ai_msg = model_with_tools.invoke(messages);
messages.append(ai_msg)

for tool_call in ai_msg.tool_calls:
    tool_result = getWeather.invoke(tool_call)
    messages.append(tool_result)

final_response = model_with_tools.invoke(messages)
print(final_response.text)
```
