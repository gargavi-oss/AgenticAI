```python
import os 
from dotenv import load_dotenv
load_dotenv()

os.environ["GROQ_API_KEY"]=os.getenv("GROQ_API_KEY")
```

```python
from langchain.chat_models import init_chat_model
model = init_chat_model("groq:openai/gpt-oss-120b")
model
```

```python
from langchain.messages import HumanMessage, SystemMessage, AIMessage

messages = [
    SystemMessage("Your are a poetry expert"),
    HumanMessage("Write me a poem on langchain")
]

responses = model.stream(messages)
```

```python
from langchain.messages import ToolMessage,AIMessage,HumanMessage
ai_msg = AIMessage(
    content=[],
    tool_calls=[{
        "name":"get_weather",
        "args": {"location":"San franciso"},
        "id":"call_123"
    }]
)
weather_result="Sunny, 72F"
tool_msg = ToolMessage(
    content=weather_result,
    tool_call_id="call_123"
)

messages=[
    HumanMessage("what is the weather in san francisco?"),
    ai_msg,
    tool_msg
]

response = model.invoke(messages)
```
